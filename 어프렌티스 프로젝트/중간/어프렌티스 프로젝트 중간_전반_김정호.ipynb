```python
# 필요한 라이브러리 임포트
import pandas as pd

# 누락된 데이터 확인
missing_data = laptop_data.isnull().sum()
print(missing_data)

# 중복된 데이터 확인
duplicate_rows = laptop_data.duplicated().sum()
print("duplicate_rows : ", duplicate_rows)

# 중복된 행 제거
laptop_data_cleaned = laptop_data.drop_duplicates()

# 중복 행 제거 후 데이터의 행 수 확인
remaining_rows = laptop_data_cleaned.shape[0]
print("remaining_rows  : ", remaining_rows)
```

```python
# 범주형 특성에 One-Hot Encoding 적용
laptop_encoded = pd.get_dummies(laptop_data_cleaned, columns=["Company", "TypeName", "OpSys", "CpuCompany", "GPU"], drop_first=True)

# 변환된 데이터의 처음 몇 행 확인
laptop_encoded.head()
```

```python
# 필요한 라이브러리 임포트
from sklearn.preprocessing import StandardScaler

# 수치형 특성 리스트
numeric_features = ["Ram", "Weight", "ClockSpeed", "Flash Storage", "HDD", "Hybrid", "SSD", "Ppi"]

# StandardScaler 객체 생성
scaler = StandardScaler()

# 수치형 특성에 Standard Scaling 적용
laptop_encoded[numeric_features] = scaler.fit_transform(laptop_encoded[numeric_features])

# 변환된 데이터의 처음 몇 행 확인
laptop_encoded.head()
```

```python
# 필요한 라이브러리 임포트
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np

# HighRam 특성 생성을 위한 사용자 정의 변환기
class HighRamFeature(BaseEstimator, TransformerMixin):
    def __init__(self, ram_column):
        self.ram_column = ram_column
        
    def fit(self, X, y=None):
        return self  # Nothing to do during fitting
    
    def transform(self, X, y=None):
        ram = X[self.ram_column]
        high_ram = (ram > 8).astype(int)
        X_copy = X.copy()
        X_copy['HighRam'] = high_ram
        return X_copy

# 사용자 정의 변환기를 사용하여 HighRam 특성 생성
high_ram_transformer = HighRamFeature(ram_column="Ram")
laptop_with_high_ram = high_ram_transformer.transform(laptop_encoded)

# 변환된 데이터의 처음 몇 행 확인
laptop_with_high_ram.head()
```

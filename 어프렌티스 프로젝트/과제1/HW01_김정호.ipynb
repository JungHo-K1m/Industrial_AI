```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
import pandas as pd
import numpy as np
import random as rnd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/dhwan/Desktop/titanic.csv")
```

```python
# 1. 생존자 중 여성이 몇 명인지 계산(과제에 명시되어있지 않아서 임의로 여성의 수를 구하였습니다.)
survived_females = df[(df['Survived'] == 1) & (df['Sex'] == 'female')].shape[0]
print(survived_females )
```

```python
# 2. 항구 "S"에서 탑승한 승객 수를 계산
embarked_S = df[df['Embarked'] == 'S'].shape[0]
print(embarked_S)
```

```python
# 3. 남성 승객 수를 계산
male_passengers = df[df['Sex'] == 'male'].shape[0]
print(male_passengers)
```

```python
# 4. 3등실에 탑승하고 지불한 요금이 40.000 미만인 승객 수를 계산
third_class_below_40 = df[(df['Pclass'] == 3) & (df['Fare'] < 40)].shape[0]
print(third_class_below_40)
```

```python
# 5. 승객 ID가 674인 승객의 이름을 찾아보세요
name_674 = df[df['PassengerId'] == 674]['Name'].values[0]
print(name_674)
```

```python
# 6. 승객들의 평균 나이를 계산
average_age = df['Age'].mean()
print(average_age)
```

```python
# 7. 생존자와 사망자의 수를 각각 계산
survived = df[df['Survived'] == 1].shape[0]
not_survived = df[df['Survived'] == 0].shape[0]
print(survived)
print(not_survived)
```

```python
# 8. 남성과 여성 승객 수를 계산
female_passengers = df[df['Sex'] == 'female'].shape[0]
print(female_passengers)
```

```python
# 9. 각 성별에 따른 생존자와 사망자의 수를 계산
survival_sex = df.groupby('Sex')['Survived'].value_counts().unstack()
print(survival_sex)
```

```python
# 10. 각 등급별로 생존한 사람과 사망한 사람의 평균 지불 요금을 계산
average_fare = df.pivot_table(index='Pclass', columns='Survived', values='Fare', aggfunc='mean')
print(average_fare)
```

```python
### II. DataFrame 데이터 시각화:

# 1. 탑승자의 생존 여부에 대한 비율을 막대 그래프로 시각화
sns.countplot(x='Survived', data=df)
plt.show()
```

```python
# 2. 나이 열의 분포를 20개의 구간(bin)으로 나누어 히스토그램으로 시각화
sns.histplot(data=df, x="Age", bins=20, kde=True)
plt.show()
```

```python
# 3. 나이와 요금 간의 산점도 시각화
sns.scatterplot(data=df, x="Age", y="Fare", hue="Survived", palette="coolwarm")
plt.show()
```

```python
# 4. 객실 등급(Pclass)별로 생존자와 사망자 수를 막대그래프로 시각화
df['Survived'] = df['Survived'].astype(str)
sns.countplot(data=df, x='Pclass', hue='Survived', palette="coolwarm")
plt.show()
```
